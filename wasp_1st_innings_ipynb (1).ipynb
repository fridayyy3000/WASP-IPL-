```python
#data just after 5h over
unique_match_ids = delivery['match_id'].unique()
data_5 = pd.DataFrame(columns=delivery.columns)

for match_id in unique_match_ids:
    match_df = delivery[delivery['match_id'] == match_id]
    desired_row = match_df[match_df['balls_bowled'] == 30]
    data_5 = data_5.append(desired_row, ignore_index=True)
```

```python
data_5
```

```python
variables = ['match_id', 'city','batting_team','bowling_team','batsman','non_striker','bowler','batsman_runs','bowler_wickets','current_score','balls_bowled','wickets','target']
```

```python
data_5 = data_5[variables]
data_5
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

data_5.fillna(0, inplace=True)

data_5 = pd.get_dummies(data_5, columns=['city', 'batting_team', 'bowling_team','batsman', 'non_striker', 'bowler'], drop_first=True)


# data_5 = data_5.drop(['batsman', 'non_striker', 'bowler'], axis=1)

X = data_5.drop(['target'], axis=1)
y = data_5['target']


X_train, X_test, y_train, y_test_5 = train_test_split(X, y, test_size=0.1, random_state=42)

#Decision Tree
dt_model_5 = DecisionTreeRegressor(random_state=42)
dt_model_5.fit(X_train, y_train)

# Random Forest Model
rf_model_5 = RandomForestRegressor(random_state=42)
rf_model_5.fit(X_train, y_train)


dt_predictions_5 = dt_model_5.predict(X_test)
rf_predictions_5 = rf_model_5.predict(X_test)
# svm_predictions = svm.predict(X_test)

dt_mae_5 = mean_absolute_error(y_test_5, dt_predictions_5)
rf_mae_5 = mean_absolute_error(y_test_5, rf_predictions_5)
# svm_mae = mean_absolute_error(y_test_5, svm_predictions)

dt_mse_5 = mean_squared_error(y_test_5, dt_predictions_5)
rf_mse_5 = mean_squared_error(y_test_5, rf_predictions_5)
# svm_mse = mean_squared_error(y_test_5, svm_predictions)

print("Decision Tree MAE:", dt_mae_5)
print("Random Forest MAE:", rf_mae_5)
# print("SVM:", svm_mae)
print("Decision Tree MSE:", dt_mse_5)
print("Random Forest MSE:", rf_mse_5)
# print("SVM:", svm_mse)
```

```python
y_preds_5 = lin_reg_5.predict(X_test_scaled)
y_preds_5
```

**ENSEMBLE **

```python
# BAGGING with MAX VOTING

from sklearn.ensemble import VotingRegressor
ensemble_5 = VotingRegressor(estimators=[
    ('tree', dt_model_5),
    ('rf', rf_model_5),
    ('xgb', xgb_5)
])
ensemble_5.fit(X_train,y_train)
y_pred_ensemble_5 = ensemble_5.predict(X_test)
y_pred_ensemble_5
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

dt_feature_importances = dt_model_5.feature_importances_
rf_feature_importances = rf_model_5.feature_importances_
xgb_feature_importances = xgb_5.feature_importances_

# Average feature importances from individual models
average_feature_importances = (dt_feature_importances + rf_feature_importances + xgb_feature_importances) / 3

# Create a DataFrame to store feature importances
feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': average_feature_importances})

# Sort the DataFrame by importance values in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Create a heatmap
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Importance', y='Feature', data=feature_importance_df, palette='viridis')
ax.set_title('Feature Importance - Ensemble (5th Over)')

# Rotate y-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

# Visualization of Predicted vs Actual Values
def plot_comparison(y_true, y_pred, title):
    plt.scatter(np.arange(len(y_true)), y_true, label='Actual Values', marker='o')
    plt.scatter(np.arange(len(y_pred)), y_pred, label='Predicted Values', marker='x')
    plt.title(title)
    plt.xlabel('Data Points')
    plt.ylabel('Values')
    plt.legend()
    plt.show()

# Assuming 'y_test_5', 'y_true_10th', 'y_test_15' are the actual values and
# 'y_pred_ensemble_5', 'ensemble_10', 'ensemble_15' are the predicted values

plot_comparison(y_test_5, y_pred_ensemble_5, 'Comparison for 5th Over')
plot_comparison(y_test_10, y_pred_ensemble_10, 'Comparison for 10th Over')
plot_comparison(y_test_15, y_pred_ensemble_15, 'Comparison for 15th Over')
```
